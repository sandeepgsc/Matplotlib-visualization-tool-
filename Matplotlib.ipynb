### Bar_plot

```python
course = ['Python','Java','.NET c#','GO']
revenue=[900,1660,891,247]
```

```python
a = np.arange(len(course))
a
```

```python
plt.bar(a,revenue,label='Revenue',color = 'red')
plt.title('Bar_plot')
plt.xticks(a,course)
plt.ylabel('Revenue')
plt.legend(loc=1 , fontsize='large')
plt.show()
```

```python
plt.barh(a,revenue,label='Revenue')
plt.title('Bar_plot')
plt.yticks(a,course)
plt.xlabel('Revenue')
plt.legend(loc=1)
plt.show()
```

## Histogram

```python
a = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(a , bins = 50) 
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist(a, bins=[80,100,125,150],rwidth=0.5, color='r')
```

```python
plt.xlabel("company")
plt.ylabel("staff")
plt.title("company status")
a = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
b = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
plt.hist([a,b], bins=[80,100,125,150], rwidth=None, color=['red','blue'],label=['men','women'],
        )
plt.legend(loc=1,fontsize='xx-large')
```

## Pie_chart

```python
x = [40,30,20,10,1]
y = ['java','python','C#','PHP','c++']
plt.pie(x,labels=y) #labels means showing 
```

```python
plt.figure(figsize = (10 , 10))
x = [40,30,20,10,1]
y = ['java','python','C#','PHP','c++']
plt.pie(x,labels=y )
```

```python
#plt.figure(figsize = (10 , 10))
x = [40,30,20,10,1]
y = ['java','python','C#','PHP','c++']
plt.pie(x,labels=y , shadow = True , autopct='%1.1f%%',radius = 1.25)
```

```python
#plt.figure(figsize = (10 , 10))
x = [40,30,20,10,1]
y = ['java','python','C#','PHP','c++']
plt.pie(x,labels=y , shadow = True , autopct='%1.1f%%',radius = 2.0,
        explode = [0,0.4,0,0,0.4])
```
